### Importación de datos



```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda.iloc[:, 2].sum()
print(ingreso_tienda1)
```

```python
ingreso_tienda2 = tienda2.iloc[:, 2].sum()
print(ingreso_tienda2)
```

```python
ingreso_tienda3 = tienda3.iloc[:, 2].sum()
print(ingreso_tienda3)
```

```python
ingreso_tienda4 = tienda4.iloc[:, 2].sum()
print(ingreso_tienda4)
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y sumar la cantidad vendida por tienda
categorias_tienda1 = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
categorias_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
categorias_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
categorias_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)

# Mostrar resultados
print(" Tienda 1 - Categorías más populares:\n", categorias_tienda1, "\n")
print(" Tienda 2 - Categorías más populares:\n", categorias_tienda2, "\n")
print(" Tienda 3 - Categorías más populares:\n", categorias_tienda3, "\n")
print(" Tienda 4 - Categorías más populares:\n", categorias_tienda4, "\n")
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_1 = tienda.iloc[:, 7].mean()


promedio_calificacion_2 = tienda2.iloc[:, 7].mean()


promedio_calificacion_3 = tienda3.iloc[:, 7].mean()


promedio_calificacion_4 = tienda4.iloc[:, 7].mean()



print(" Tienda 1 - Calificación:\n", promedio_calificacion_1, "\n")
print(" Tienda 2 - Calificación:\n", promedio_calificacion_2, "\n")
print(" Tienda 3 - Calificación:\n", promedio_calificacion_3, "\n")
print(" Tienda 4 - Calificación:\n", promedio_calificacion_4, "\n")
```

# 4. Productos más y menos vendidos

```python
productos_t1 = tienda.groupby('Producto')['Producto'].count().sort_values(ascending=False)
productos_t2 = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)
productos_t3 = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)
productos_t4 = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)
```

```python
# Mostrar los más y menos vendidos por tienda
print("+ Más vendidos - Tienda 1:\n", productos_t1.head(5), "\n")
print("- Menos vendidos - Tienda 1:\n", productos_t1.tail(5), "\n")

print("+ Más vendidos - Tienda 2:\n", productos_t2.head(5), "\n")
print("- Menos vendidos - Tienda 2:\n", productos_t2.tail(5), "\n")

print("+ Más vendidos - Tienda 3:\n", productos_t3.head(5), "\n")
print("- Menos vendidos - Tienda 3:\n", productos_t3.tail(5), "\n")

print("+ Más vendidos - Tienda 4:\n", productos_t4.head(5), "\n")
print("- Menos vendidos - Tienda 4:\n", productos_t4.tail(5), "\n")
```

# 5. Envío promedio por tienda

```python
envio_promedio_1 = tienda.iloc[:, 3].mean()


envio_promedio_2 = tienda2.iloc[:, 3].mean()


envio_promedio_3 = tienda3.iloc[:,3].mean()


envio_promedio_4 = tienda4.iloc[:, 3].mean()
```

```python
print(" Tienda 1 - Envío promedio:\n", envio_promedio_1, "\n")
print(" Tienda 2 - Envío promedio:\n", envio_promedio_2, "\n")
print(" Tienda 3 - Envío promedio:\n", envio_promedio_3, "\n")
print(" Tienda 4 - Envío promedio:\n", envio_promedio_4, "\n")
```

```python
import matplotlib.pyplot as plt

import matplotlib.ticker as ticker




# Nombres y valores
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# Crear gráfico
plt.bar(tiendas, ingresos, color='GREEN')
plt.title('💰 Ingresos Totales por Tienda')
plt.ylabel('Ingreso ($)')
plt.xlabel('Tienda')
plt.tight_layout()

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Mostrar el gráfico

plt.show()

tiendas
ingresos

```

```python
def crear_grafico_pastel(data, titulo):
    plt.figure(figsize=(7, 7))  # Ajusta el tamaño de la figura
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=40)
    plt.title(titulo)
    plt.show()

# Crear gráficos de pastel para cada tienda
crear_grafico_pastel(categorias_tienda1, 'Tienda 1 - Categorías más populares')
crear_grafico_pastel(categorias_tienda2, 'Tienda 2 - Categorías más populares')
crear_grafico_pastel(categorias_tienda3, 'Tienda 3 - Categorías más populares')
crear_grafico_pastel(categorias_tienda4, 'Tienda 4 - Categorías más populares')
```

```python

calificaciones= [promedio_calificacion_1, promedio_calificacion_2, promedio_calificacion_3, promedio_calificacion_4]

# Crear gráfico





plt.figure(figsize=(8, 6))  # Ajustar el tamaño de la figura
plt.scatter(tiendas, calificaciones, s=100, c='blue', marker='o')

# Añadir etiquetas a los ejes
plt.xlabel('Tiendas')
plt.ylabel('Promedio de Calificación')

# Añadir un título
plt.title('Promedio de Calificación por Tienda')

# Mostrar el gráfico
plt.grid(True)  # Añadir una cuadrícula para facilitar la lectura
plt.show()
```
